```python
import pandas as pd
import numpy as np
```

```python
raw_us = pd.read_csv("../resources/us-maps/us-state-counties.tsv", sep="\t")
raw_us.iloc[0]
```

```python
us = pd.DataFrame(index=raw_us.index)
us["country"] = "United States"
us["stateprovince"] = raw_us["state"]
us["county"] = raw_us["county"]
us.head()
```

```python
df = us.copy()
df["kingdom"] = "plantae"
df["phylum"] = "tracheophyta"
df["family"] = "sapindaceae"
df["genus"] = "acer"
df["specificepithet"] = "saccharum"
sugarmaple = df
sugarmaple.head()
```

```python
df = us.copy()
df["kingdom"] = "plantae"
df["phylum"] = "tracheophyta"
df["family"] = "fabaceae"
df["genus"] = "amorpha"
df["specificepithet"] = "canescens"
leadplant = df
leadplant.head()
```

```python
df = us.copy()
df["kingdom"] = "animalia"
df["phylum"] = "chordata"
df["family"] = "dasypodidae"
df["genus"] = "dasypus"
df["specificepithet"] = "novemcinctus"
armadillo = df
armadillo.head()
```

```python
df = us.copy()
df["kingdom"] = "animalia"
df["phylum"] = "chordata"
df["family"] = "picidae"
df["genus"] = "leuconotopicus"
df["specificepithet"] = "albolarvatus"
woodpecker = df
woodpecker.head()
```

```python
pd.concat([sugarmaple, leadplant, armadillo, woodpecker]).to_csv("../results/us-maps/all.tsv", sep="\t", index=False)
```
